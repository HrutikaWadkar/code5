```python
import numpy as np
import matplotlib.pyplot as plt

# Define mean and standard deviation
mean = 0
std_dev = 1

# Generate random data from a normal distribution
data = np.random.normal(mean, std_dev, 1000)

# Create a range of values for plotting the normal distribution curve
x = np.linspace(min(data), max(data), 1000)

# Compute the normal distribution probability density function (PDF)
pdf = (1/(std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev)**2)

# Plotting the histogram of the generated data
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Generated Data')

# Plotting the normal distribution curve
plt.plot(x, pdf, label='Normal Distribution Curve', color='b', linewidth=2)

# Adding titles and labels
plt.title('Normal Distribution Curve')
plt.xlabel('Data Values')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()
```
